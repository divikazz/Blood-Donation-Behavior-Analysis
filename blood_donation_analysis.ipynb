## 4. EDA

```python
#how often people do the blood donation
import matplotlib.pyplot as plt

df["Frequency"].hist()

plt.title("Distribution of Blood Donation Frequency")
plt.xlabel("Number of Donations")
plt.ylabel("Count")
plt.show()
```

```python
#how long since the last donation
df["Recency"].hist()

plt.title("Distribution of Recency")
plt.xlabel("Months Since Last Donation")
plt.ylabel("Count")
plt.show()
```

```python
#to see donation's pattern
plt.scatter(df["Time"], df["Frequency"])

plt.title("Donation Frequency vs Donor History")
plt.xlabel("Months Since First Donation")
plt.ylabel("Total Donations")
plt.show()
```

```python
df["Donated"].value_counts()
```

```python
df["Donated"].value_counts().plot(kind="bar")

plt.title("Repeat Donation")
plt.xlabel("Donated Again")
plt.ylabel("Number of Donors")
plt.show()
```

## 5. Key Insights
1. Most donors donate only a few times.  
   The distribution of donation frequency shows that the majority of donors have donated a small number of times.

2. A small group of donors contributes a large portion of total donations.  
   Frequent donors represent a smaller segment but play a critical role in maintaining blood supply.

3. Donors with longer donation history tend to donate more frequently.  
   The relationship between donation frequency and time since first donation indicates that long-term donors are more active.

4. Some donors have not donated for a long time.  
   The recency distribution suggests that many donors have not donated recently, which presents an opportunity for targeted donation campaigns.

5. Identifying frequent donors can help improve blood donation programs.  
   Focusing on donor retention strategies could increase the stability of blood supply.

## Conclusion

This analysis explored blood donation behavior using historical donation data.
The results show that most donors donate only a few times, while a smaller group of regular donors contributes a significant portion of the blood supply.

Donors who have been donating for a longer period tend to donate more frequently, indicating the importance of maintaining long-term donor relationships. Additionally, many donors have not donated recently, suggesting potential opportunities for re-engagement campaigns.

Understanding donor behavior patterns can help blood donation organizations design more effective recruitment and retention strategies to maintain a stable blood supply.
